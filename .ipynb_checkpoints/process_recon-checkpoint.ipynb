```python
import pandas as pd

#function to find the line number of different strings
def get_line_number(string, data):
    string_index = data.index(string)
    lines = 0
    i = 0
    while i < string_index:
        if (data[i] == '\n'):
            lines += 1
        i += 1
    return lines+1

    
all_data = open('recon_in.txt').read()
d0_pos_start = get_line_number('D0-POS', all_data)
d1_trn_start = get_line_number('D1-TRN', all_data)
d1_pos_start = get_line_number('D1-POS', all_data)

d0_pos = pd.read_csv('recon_in.txt', delimiter=' ', names=['stock', 'shares_0'], 
                     skiprows=d0_pos_start, nrows=(d1_trn_start-d0_pos_start-2)) 
d1_trn = pd.read_csv('recon_in.txt', delimiter=' ', names=['stock', 'action', 'shares', 'value'], 
                     skiprows=d1_trn_start, nrows=(d1_pos_start-d1_trn_start-2))
d1_pos = pd.read_csv('recon_in.txt', delimiter=' ', names=['stock', 'shares_1'], 
                     skiprows=d1_pos_start)

#set up a dictionary of possible transaction types we might see
#if it is something that adds, assign value of 1; if it subtracts, assign -1
cash_actions = {
    'SELL': 1,
    'DEPOSIT': 1,
    'DIVIDEND': 1,
    'BUY': -1,
    'FEE': -1
}

#manipulate d1_trn into a table that shows the stock name and the change
    #add Cash in using the 4th value
change_obj = {}
cash_change = 0

for pt in d1_trn.values:
    if pt[0] in change_obj:
        change_obj[pt[0]] += -1*cash_actions[pt[1]]*pt[2]
    else:
        change_obj[pt[0]] = -1*cash_actions[pt[1]]*pt[2]
    cash_change += cash_actions[pt[1]]*pt[3]

change_obj['Cash'] = cash_change
delta = pd.DataFrame(change_obj.items(), columns=['stock', 'change'])


# use outer joins to merge the differet tables togehter properly
full_merge = pd.merge((pd.merge(d0_pos, delta, on='stock', how='outer').fillna(0)), d1_pos, on='stock', how='outer').fillna(0)

#Calculate the recon_out using data in dataframe
full_merge['recon_out'] = full_merge['shares_1'] - full_merge['shares_0'] - full_merge['change']
#Grab the stock name and the recon_out value from the table where recon_out != 0
recon_out_df = full_merge[full_merge['recon_out'] != 0][['stock', 'recon_out']]
#Turn dataframe back into a space-separated file
recon_out_df.to_csv('recon_out.txt', sep=' ', index=False, header=False)
```
